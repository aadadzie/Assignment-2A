## Prediction

### Building Simple Models using;
* Mean
* Median
* Max
* Min
* Standard Deviation

### In each of the above, we will;
* Assign target variable to y
* Model the target variable
* Assess the model performance

```python
#defining functions for model performance

def mse(y, y_pred):
    mean_squared_error = ((y - y_pred) ** 2).sum() / len(y)
    
    return mean_squared_error

def mae(y, y_pred):
    mean_absolute_error = np.absolute(y_pred - y).sum() / len(y)
    
    return mean_absolute_error
```

#### Assigning Target variable to y

```python
y = housedata['price']
y.head()
```

#### Modeling Target variable

```python
# Model 1 - Mean
y_pred = y.mean()
model_1_mse = mse(y,y_pred); model_1_mae = mae(y,y_pred);
print(y_pred)
print('Mean Model_mse: ', model_1_mse)
print('Mean Model_mae: ', model_1_mae)
```

```python
# Model 2 - Median
y_pred = y.median()
model_2_mse = mse(y,y_pred); model_2_mae = mae(y,y_pred);
print(y_pred)
print('Median Model_mse: ', model_2_mse)
print('Median Model_mae: ', model_2_mae)
```

```python
# Model 3 - Max
y_pred = y.max()
model_3_mse = mse(y,y_pred); model_3_mae = mae(y,y_pred);
print(y_pred)
print('Max Model_mse: ', model_3_mse)
print('Max Model_mae: ', model_3_mae)
```

```python
# Model 4 - Min
y_pred = y.min()
model_4_mse = mse(y,y_pred); model_4_mae = mae(y,y_pred);
print(y_pred)
print('Min Model_mse: ', model_4_mse)
print('Min Model_mae: ', model_4_mae)
```

```python
# Model 5 - Standard Deviation
y_pred = y.std()
model_5_mse = mse(y,y_pred); model_5_mae = mae(y,y_pred);
print(y_pred)
print('Std Model_mse: ', model_5_mse)
print('Std Model_mae: ', model_5_mae)
```

```python
#Plotting mse errors for simple models;
errors = {'model_1_mse': model_1_mse,
          'model_2_mse': model_2_mse,
          'model_3_mse': model_3_mse,
          'model_4_mse': model_4_mse, 
          'model_5_mse': model_5_mse}

np.log(pd.Series(errors)).plot(kind='bar', figsize=(10, 5), rot=0)
```

```python
#Plotting mae errors for simple models;
errors = {'model_1_mae': model_1_mae,
          'model_2_mae': model_2_mae,
          'model_3_mae': model_3_mae,
          'model_4_mae': model_4_mae, 
          'model_5_mae': model_5_mae}

np.log(pd.Series(errors)).plot(kind='bar', figsize=(10, 5), rot=0)
```

#### Using median price of houses from each city to predict prices coming from that city

```python
city_median = housedata.groupby('city')['price'].median()
city_median.head()
```

```python
# to pick the median prices(values) from each city. ie:city_median dataframe and assign/wrap values to the original dataframe(housedata['city']) 
# this allows y and y_pred to be of equal vectors.
city_median.loc[housedata['city']]                        # is the same as city_median[housedata['city']]
```

```python
# then assigning the predicted values which is in a pandas series form and of equal length/vector as y actual
# so that we can assess our model performance.

y_pred_city = city_median.loc[housedata['city']].values
y_pred_city
```

```python
model_6_mse = mse(y,y_pred_city); model_6_mae = mae(y,y_pred_city);
print('City Model_mse: ', model_6_mse)
print('City Model_mae: ', model_6_mae)
```

```python
# plotting the errors from city model predictor;
errors = {'City Model_mse':model_6_mse, 'City Model_mae':model_6_mae}
np.log(pd.Series(errors)).plot(kind='bar', rot=0)
```
